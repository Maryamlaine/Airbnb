### Table Review

```python
table_review =  pd.merge(df, 
                        df_2,
                        how="inner",
                        left_on = 'id',
                        right_on= 'listing_id')
```

```python
df_review = table_review[["listing_id", "number_of_reviews", "review_scores_rating", "review_scores_accuracy", "review_scores_checkin",
               "review_scores_communication", "review_scores_location", "review_scores_value", "comments"]]
```

```python
df_review['review_id'] = [i for i in range(1, len(df_review) + 1)]
```

```python
df_review = df_review[["review_id","listing_id", "number_of_reviews", "review_scores_rating", "review_scores_accuracy", "review_scores_checkin",
               "review_scores_communication", "review_scores_location", "review_scores_value", "comments"]]
df_review.columns = ["review_id","property_id", "number_of_reviews", "review_scores_rating", "review_scores_accuracy", "review_scores_checkin",
               "review_scores_communication", "review_scores_location", "review_scores_value", "comments"]
```

```python
df_review[['review_scores_rating', 'review_scores_accuracy', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']] = df_review[['review_scores_rating', 'review_scores_accuracy', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']].astype('Int64')
```

```python
df_review['comments'] = df_review['comments'].str.replace('"', '')
df_review['comments'] = df_review['comments'].str.replace("'", '')
```

```python
df_review
```

### Table Host

```python
df_host = df[["host_id", "host_url", "host_name", "host_since","host_is_superhost", "host_response_time",
              "host_acceptance_rate"]]
```

```python
df_host.drop_duplicates("host_id", keep='last', inplace=True)
```

```python
df_host['host_is_superhost'] = df_host['host_is_superhost'].map({'t': 1, 'f': 0}).astype('Int64')
```

```python
df_host
```

### Table Price

```python
df_price = df[["price", "weekly_price", "monthly_price","cleaning_fee", "security_deposit"]]
df_price['price_id'] = [i for i in range(1, len(df_price) + 1)]
df_price = df_price[['price_id',"price", "weekly_price", "monthly_price","cleaning_fee", "security_deposit"]]
df_price.rename(columns={'price':'daily_price'}, inplace=True)
df_price = df_price.replace(',','', regex=True)
df_price['daily_price'] = df_price['daily_price'].str.replace('$', '').astype(float)
df_price['weekly_price'] = df_price['weekly_price'].str.replace('$', '').astype(float)
df_price['monthly_price'] = df_price['monthly_price'].str.replace('$', '').astype(float)
df_price['cleaning_fee'] = df_price['cleaning_fee'].str.replace('$', '').astype(float)
df_price['security_deposit'] = df_price['security_deposit'].str.replace('$', '').astype(float)
df_price 
```

### Table Neighborhood

```python
df_neighborhood = df[["latitude", "longitude", "city", "state", "zipcode", "access", "transit"]]
df_neighborhood.drop_duplicates(["latitude", "longitude"], keep='last', inplace=True)
```

```python
df_neighborhood['neighborhood_id'] = [i for i in range(1, len(df_neighborhood) + 1)]
df_neighborhood = df_neighborhood[["neighborhood_id", "latitude", "longitude","city", "state", 
                                "zipcode", "access", "transit"]]
df_neighborhood['access'] = df_neighborhood['access'].str.strip("''")
df_neighborhood
```

### Table Availability 

```python
df_availability = df[["availability_30", "availability_60", "availability_90","availability_365"]]
```

```python
df_availability.shape
```

```python
df_availability['availability_id'] = [i for i in range(1, len(df_availability) + 1)]
df_availability = df_availability[["availability_id","availability_30", "availability_60", "availability_90","availability_365"]]
df_availability
```

### Table Amenity

```python
table_amenities = df[['id', 'amenities']]
table_amenities['amenities'] = table_amenities['amenities'].str.strip('{}')
```

```python
table_amenities['wireless_Internet'] = df.amenities.str.findall("(Wireless Internet)").apply(', '.join, 1)
table_amenities['air_conditioning'] = df.amenities.str.findall("(Air conditioning)").apply(', '.join, 1)
table_amenities['pool'] = df.amenities.str.findall("(Pool)").apply(', '.join, 1)
table_amenities['self_check_in'] = df.amenities.str.findall("(Self Check-In)").apply(', '.join, 1)
table_amenities['private_entrance'] = df.amenities.str.findall("(Private entrance)").apply(', '.join, 1)
table_amenities['breakfast'] = df.amenities.str.findall("(Breakfast)").apply(', '.join, 1)
table_amenities['free_parking_on_premises'] = df.amenities.str.findall("(Free parking on premises)").apply(', '.join, 1)
table_amenities['tv'] = df.amenities.str.findall("(TV)").apply(', '.join, 1)
table_amenities['gym'] = df.amenities.str.findall("(Gym)").apply(', '.join, 1)
table_amenities['hot_tub'] = df.amenities.str.findall("(Hot tub)").apply(', '.join, 1)
table_amenities['indoor_fireplace'] = df.amenities.str.findall("(Indoor fireplace)").apply(', '.join, 1)
table_amenities['heating'] = df.amenities.str.findall("(Heating)").apply(', '.join, 1)
table_amenities['family_kid_friendly'] = df.amenities.str.findall("(Family/kid friendly)").apply(', '.join, 1)
table_amenities['washer'] = df.amenities.str.findall("(Washer)").apply(', '.join, 1)
table_amenities['dryer'] = df.amenities.str.findall("(Dryer)").apply(', '.join, 1)
table_amenities['hangers'] = df.amenities.str.findall("(Hangers)").apply(', '.join, 1)
table_amenities['iron'] = df.amenities.str.findall("(Iron)").apply(', '.join, 1)
table_amenities['shampoo'] = df.amenities.str.findall("(Shampoo)").apply(', '.join, 1)
table_amenities['hair_dryer'] = df.amenities.str.findall("(Hair dryer)").apply(', '.join, 1)
table_amenities['high_chair'] = df.amenities.str.findall("(High chair)").apply(', '.join, 1)
table_amenities['crib'] = df.amenities.str.findall("(crib)").apply(', '.join, 1)
table_amenities['carbon_monoxide_detector'] = df.amenities.str.findall("(Carbon monoxide detector)").apply(', '.join, 1)
table_amenities['smoke_detector'] = df.amenities.str.findall("(Smoke detector)").apply(', '.join, 1)
table_amenities['babysitter_recommendations'] = df.amenities.str.findall("(Babysitter recommendations)").apply(', '.join, 1)
table_amenities['pets_allowed'] = df.amenities.str.findall("(Pets allowed)").apply(', '.join, 1)
table_amenities['first_aid_kit'] = df.amenities.str.findall("(First aid kit)").apply(', '.join, 1)
```

```python
table_amenities.drop('amenities', axis=1, inplace=True)
table_amenities = table_amenities.replace(r'^\s*$', np.nan, regex=True)
df_amenities = table_amenities.notnull().astype('int')
df_amenities.rename(columns={'id': 'amenity_id'}, inplace=True)
df_amenities['amenity_id'] = df_neighborhood['neighborhood_id'].values
df_amenities
```

### Table Property

```python
df_property_1 = pd.merge(df, 
                        df_neighborhood,
                        how='left',
                        on = ["latitude", "longitude"])
```

```python
df_property = df_property_1[["id", "host_id", "neighborhood_id", "name", "description", "property_type", "bedrooms", 
                             "bathrooms", "accommodates", "beds", "bed_type", "room_type", "minimum_nights",
                             "maximum_nights", "extra_people"]]
df_property['price_id'] = [i for i in range(1, len(df_property) + 1)]
df_property['availability_id'] = [i for i in range(1, len(df_property) + 1)]
df_property['amenity_id'] = [i for i in range(1, len(df_property) + 1)]
df_property = df_property[["id", "host_id", "neighborhood_id", "price_id", "availability_id",'amenity_id', "name", 
                           "description", "property_type", "bedrooms", "bathrooms", "accommodates", "beds", "bed_type",
                           "room_type", "minimum_nights", "maximum_nights", "extra_people"]]
```

```python
df_property.rename({'id': 'property_id', 'name': 'listing_name', 'description': 'listing_description', 'extra_people': 'extra_people_cost'}, axis=1, inplace=True)
df_property['extra_people_cost'] = df_property['extra_people_cost'].str.strip('$')
df_property['listing_description'] = df_property['listing_description'].str.replace('"', '')
df_property['listing_description'] = df_property['listing_description'].str.replace("'", '')
df_property
```
